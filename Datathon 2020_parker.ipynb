```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pp
import seaborn as sns
data = pd.read_csv("OneDrive/datathon/complaints.csv")
```

```python
data.head()
for col in data.columns:
    print(col)
```

```python
## Preprocessing text data
```

```python
def preprocess(ReviewText):
    ReviewText = ReviewText.str.replace("(<br/>)", "")
    ReviewText = ReviewText.str.replace('(<a).*(>).*(</a>)', '')
    ReviewText = ReviewText.str.replace('(&amp)', '')
    ReviewText = ReviewText.str.replace('(&gt)', '')
    ReviewText = ReviewText.str.replace('(&lt)', '')
    ReviewText = ReviewText.str.replace('(\\xa0)', ' ')
    ReviewText = ReviewText.str.lower()
    return ReviewText
data['Company'] = preprocess(data['Company'])
data['Company public response'] = preprocess(data['Company public response'])
data['Consumer complain narrative'] = preprocess(data['Consumer complaint narrative'])
```

```python
data['Date received'] = pd.to_datetime(data['Date received'])
data['Date sent to company'] = pd.to_datetime(data['Date sent to company'])
```

```python
## Create pivots
```

```python
prod_subprod = data.groupby(['Product', 'Sub-product'])
issue_subissue = data.groupby(['Issue', 'Sub-issue'])
date_issue = data.groupby(['Date received', 'Issue'])
```

```python
## Product
pivot = data.pivot_table(index = ['Product'], values = ['Complaint ID'], aggfunc = 'count')
pivot.sort_values('Complaint ID', ascending = False).head(10)
```

```python
## product sub product
pivot2 = data.pivot_table(index = ['Product', 'Sub-product'], values = ['Complaint ID'], aggfunc = 'count')
pivot2.sort_values('Complaint ID', ascending = False).head(10)
```

```python
## Issue
pivot3 = data.pivot_table(index = ['Issue'], values = ['Complaint ID'], aggfunc = 'count')
pivot3.sort_values('Complaint ID', ascending = False).head(10)
```

```python
## Issue & Sub-issue
pivot4 = data.pivot_table(index = ['Issue', 'Sub-issue'], values = ['Complaint ID'], aggfunc = 'count')
pivot4.sort_values('Complaint ID', ascending = False).head(10)
```
